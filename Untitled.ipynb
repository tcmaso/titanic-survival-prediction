# Data preprocessing pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class NewFeatures(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        X['Cabin_group'] = X['Cabin'].apply(lambda x: x[0] if pd.notnull(x) else 'Unknown')
        X['Cabin_group'][X['Cabin_group']=='T'] = 'A' # T cabin close to a
        
        X['Cabin_group'] = OrdinalEncoder(categories=[['A','B','C','D','E','F','G','Unknown']]).fit_transform(X[['Cabin_group']])
        
        X['Title'] = X['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
        X['Title'] = X['Title'].replace(['Mr', 'Master'], 'Mr/Master')
        X['Title'] = X['Title'].replace(['Miss', 'Mrs','Ms', 'Mlle', 'Lady', 'Mme', 'the Countess', 'Dona'], 'Miss/Mrs/Ms')
        X['Title'] = X['Title'].replace(['Dr', 'Col', 'Major', 'Jonkheer', 'Capt', 'Sir', 'Don', 'Rev'], 'Dr/Military/Noble/Clergy')
        
        X = X.join(pd.DataFrame(OneHotEncoder().fit_transform(X[['Title']]).toarray(), index=X.index))
        
        X.drop(['Cabin', 'Name', 'Title'], axis=1, inplace=True)
        return X
```

```python
preprocessing_num_pipeline = Pipeline(steps=[
    ('fill_na', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
preprocessing_cat_pipeline = Pipeline(steps=[
    ('fill_na', SimpleImputer(strategy='most_frequent')),
    ('one_hot_encode', OneHotEncoder())
])
```

```python
preprocessing_ord_pipeline = Pipeline(steps=[
    ('fill_na', SimpleImputer(strategy='most_frequent')),
    ('ordinal_encode', OrdinalEncoder())
])
```

```python
num_columns = ['Age', 'Fare']
list(num_columns)
```

```python
ord_columns = ['Pclass', 'SibSp', 'Parch']
ord_columns
```

```python
cat_columns = ['Sex', 'Embarked']
cat_columns
```

```python
preprocessing_pipeline = ColumnTransformer([
    ('feature_engineering', NewFeatures(), ['Cabin', 'Name']),
    ('num_pipeline', preprocessing_num_pipeline, num_columns),
    ('ordinal_pipeline', preprocessing_ord_pipeline, ord_columns),
    ('cat_pipeline', preprocessing_cat_pipeline, cat_columns),
])
```

# Training algorithms

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.tree import ExtraTreeClassifier, DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
import time
from sklearn.metrics import f1_score, confusion_matrix
```

```python
algorithms = [RandomForestClassifier(), AdaBoostClassifier(), GradientBoostingClassifier(), BaggingClassifier(),
             ExtraTreeClassifier(), DecisionTreeClassifier(), KNeighborsClassifier(), LogisticRegression(),
             SGDClassifier(), SVC(), XGBClassifier()]
```

```python
def prepare_model(X_train, y_train, algorithm):
    model = Pipeline(steps=[
        ('preprocessing', preprocessing_pipeline),
        ('model', algorithm)
    ])
    model.fit(X_train, y_train)
    return model
```

```python
names = []
times = []
f1s = []
cms = []

for algorithm in algorithms:
    name = type(algorithm).__name__
    print(name)
    names.append(name)
    
    start_time = time.time()
    model = prepare_model(X_train, y_train, algorithm)
    pred = model.predict(X_val)
    end_time = time.time()
    time_duration = end_time - start_time
    times.append(time_duration)
    print(time_duration)
    
    f1 = f1_score(y_val, pred)
    f1s.append(f1)
    cm = confusion_matrix(y_val, pred)
    cms.append(cm)
```

```python
results_dict = {'Name':names, 'Time':times, 'F1':f1s, 'CM':cms}
```

```python
pd.DataFrame(results_dict).sort_values(by='F1', ascending=False)
```

```python
model = prepare_model(X_train, y_train, XGBClassifier())
pred = model.predict(X_test)
X_test['Survived'] = pred
X_test
```
